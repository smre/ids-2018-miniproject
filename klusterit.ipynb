```python
import pandas as pd
import geopandas as gp
from shapely.geometry import Point
import json
import numpy as np


df = pd.read_csv("services_better.csv")
```

```python
df.dtypes
```

```python
df = df[df.location.notnull()]
```

```python
df.location = df.location.apply(lambda x: x.replace("'", "\""))
df.location = df.location.apply(lambda x: json.loads(x))
df['Coordinates'] = df.location.apply(lambda x: x['coordinates'])
df['Coordinates'] = df.Coordinates.apply(Point)
```

```python
gdf = gp.GeoDataFrame(df, geometry='Coordinates')
```

```python
coordinates = np.array(list(gdf['Coordinates'].apply(lambda x: list(x.coords[0]))))
coordinates
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 15, n_init=10)
kmeans.fit(coordinates)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
clusters = {}
for label in kmeans.labels_:
    if label not in clusters:
        clusters[label] = 1
    else:
        clusters[label] += 1
clusters
```
